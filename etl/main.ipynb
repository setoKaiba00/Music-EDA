```python
import requests
import pandas as pd
import os

# 1. Appeler l'API Deezer
url = "https://api.deezer.com/chart/0/tracks"
response = requests.get(url)
data = response.json()

# 2. Transformer en DataFrame
df = pd.DataFrame(data["data"])

# 3. Afficher les 5 premi√®res lignes
df.head()


output_path = os.makedirs("output", exist_ok=True)

df.to_csv("output/tracks.csv", index=False)
df.to_csv(output_path, index=False)

print(f"Saved {len(df)} rows to {output_path}")
```
