```python
import requests
import pandas as pd
import os

all_data = []
url = "https://api.deezer.com/chart/0/tracks"

while url:
    print("Fetching:", url)
    response = requests.get(url)
    data = response.json()
    
    # Append current page
    all_data.extend(data["data"])
    
    # Check if there is a next page
    url = data.get("next")   # if 'next' doesn't exist → returns None → loop stops

# Convert to DataFrame
df = pd.DataFrame(all_data)

# 2. Create output folder
#os.makedirs("output", exist_ok=True)

# 3. Save CSV
#df.to_csv("output/tracks.csv", index=False)

#print("Saved CSV successfully!")
```

```python
#removing eval function
df["artist_id"] = df["artist"].apply(lambda x : x["id"] if pd.notnull(x) else None)
df["artist_name"] = df["artist"].apply(lambda x : x["name"] if pd.notnull(x) else None)
#extract album ad/title
df["album_id"] = df["album"].apply(lambda x : x["id"] if pd.notnull(x) else None)
df["album_title"] = df["album"].apply(lambda x : x["title"] if pd.notnull(x) else None)
#drop album/artist columns
df.drop(columns=["artist","album"], inplace=True)
```

```python
#fixing dataTypes
df["duration"] = pd.to_numeric(df["duration"], errors="coerce").astype("Int64")
df["rank"] = pd.to_numeric(df["rank"], errors="coerce").astype("Int64")
df["position"] = pd.to_numeric(df["position"], errors="coerce").astype("Int64")
df["artist_id"] = pd.to_numeric(df["artist_id"], errors="coerce").astype("Int64")
df["album_id"] = pd.to_numeric(df["album_id"], errors="coerce").astype("Int64")

df["explicit_lyrics"] = df["explicit_lyrics"].astype(bool)
```

```python
import unicodedata

df["title"] = df["title"].apply(lambda x: unicodedata.normalize("NFKD", str(x)).strip())
df["artist_name"] = df["artist_name"].apply(lambda x: unicodedata.normalize("NFKD", str(x)).strip())
df["album_title"] = df["album_title"].apply(lambda x: unicodedata.normalize("NFKD", str(x)).strip())
df["type"] = df["type"].apply(lambda x: unicodedata.normalize("NFKD", str(x)).strip())
```

```python
#calculate the % of null values on each columns
print("percentage of null values after droping columns")
print((df.isnull().mean()*100).sort_values(ascending=False))
print("percentage of duplicates after droping columns")
print(f"{df.duplicated().mean()*100} duplicates")
```

```python

df["duration_min"] = (df["duration"] / 60).round(2)
```

```python
df = df[(df["duration"] >= 20) & (df["duration"] <= 900)]
df = df[df["position"] > 0]
```

```python
artist_counts = df["artist_name"].value_counts()
df["artist_popularity"] = df["artist_name"].map(artist_counts)
```
